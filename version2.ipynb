```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("train.csv")

# Convert relevant columns to numeric, coercing errors
num_features = [
    'Time_spent_Alone', 
    'Social_event_attendance', 
    'Going_outside',
    'Friends_circle_size', 
    'Post_frequency'
]

for col in num_features:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows with missing values in numerical columns for visualization
df_viz = df.dropna(subset=num_features)

# Set up the plot style
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# Create subplots for KDE plots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 12))
axes = axes.flatten()

for i, col in enumerate(num_features):
    sns.kdeplot(data=df_viz, x=col, hue='Personality', fill=True, ax=axes[i], palette="flare")
    axes[i].set_title(f"Distribution of {col} by Personality", fontsize=12)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Density")

# Remove the last empty subplot
fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```
