```python
num = train.select_dtypes(include=[np.number]).columns.tolist()
numT = test.select_dtypes(include=[np.number]).columns.tolist()

colm = train.select_dtypes(include=[object]).columns.tolist()
colmt = test.select_dtypes(include=[object]).columns.tolist()
```

```python
df_encoded = train[colm].copy()
df_encodedT = test[colmt].copy()
label_encoders = {}

for col in colm:
    df_encoded[col] = df_encoded[col].fillna("Missing")
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le
    
    
for col in colmt:
    df_encodedT[col] = df_encodedT[col].fillna("Missing")
    le = label_encoders[col]
    df_encodedT[col] = le.transform(df_encodedT[col])
    
```

```python
train = pd.concat([train[num], df_encoded], axis=1)
test = pd.concat([test[numT], df_encodedT], axis=1) 
```

```python
train = pd.DataFrame(scalar.fit_transform(train), columns=train.columns)    
test = pd.DataFrame(scalar.fit_transform(test), columns=test.columns)   
```

```python
imputer = KNNImputer(n_neighbors=5, weights='uniform')
train[train.columns] = imputer.fit_transform(train)
test[test.columns] = imputer.fit_transform(test)
```
